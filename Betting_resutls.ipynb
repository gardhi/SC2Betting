```python
import pandas as pd
import os
import datetime
import json



def dateFromLine(date_text):
    try:     
        return datetime.datetime.strptime(date_text, '%d %b %Y, %X') 
    except:
        return False

    
    
bets_dir = 'C:\\Users\\GardH\\Documents\\Betting'
content = []

for file in os.listdir(bets_dir):
    
    if 'settled' in file:
        
        with open(bets_dir + '\\' + file) as fin:
            
            content += fin.readlines()

            
                    
entries = dict()      
new_result = False

for i in range(0, len(content)):
        
    line = content[i].strip()
    
    if len(line) == 0: continue
    
    new_date = dateFromLine(line)
    
    if new_date:
        
        entry_date = line
        entries[entry_date] = dict()
        new_result = True
    
    elif new_result:
        
        entries[entry_date]['Result'] = line
        new_result = False
        new_bet = True
    
    elif new_bet:

        if 'odds' in line.lower():

            entries[entry_date]['Bet'] = line.split(':')[1]

        else:

            entries[entry_date]['Bet'] = line
    
        new_bet = False
        
    else:
        
        line = line.strip().split(':')
        
        if len(line) > 1:
            
            entries[entry_date][line[0]] = line[1].split('kr')[0].split('.')[0].strip()
        
        else:
            
            line = line[0].split('-')
            
            entries[entry_date]['Players'] = line
            
json_data = json.dumps(entries)
df = pd.read_json(json_data, orient='index')
df.fillna(0, inplace =True)
```
